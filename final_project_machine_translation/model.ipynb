```python
# import dependencies
import string
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
from sklearn.model_selection import train_test_split
from jiwer import wer
from numpy import genfromtxt

import tensorflow as tf
from tensorflow import keras
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, LSTM, RepeatVector, Dense 
from keras import optimizers
from keras.callbacks import Callback

import warnings
warnings.filterwarnings('ignore')
```

# Step 1
### Uploading and reading the data

```python
def read_txt(file_name_fr, file_name_en, number_columns):
    # reading the data
    data_fr = np.loadtxt(file_name_fr, dtype=str, delimiter = "\t")
    data_en = np.loadtxt(file_name_en, dtype=str, delimiter = "\t")
    data = np.column_stack((data_fr,data_en))

    # taking the first number_columns lines
    data = data[:number_columns,:]
    return data

def read_csv(file_name, number_columns):
    # reading the data
    df = pd.read_csv(file_name) 
    data = df.to_numpy()
    
    # taking the first number_columns lines
    data = data[:number_columns,:]
    return data

# dataset 1
data = read_txt("vocab_fr.txt", "vocab_en.txt", 1000)
# print(data)

# dataset 2
# data = read_csv("data1.csv", 1000)
# print(data)
```

# Step 2
### Cleaning and Preprocessing the data

```python
# remove punctuation
data[:,0] = [s.translate(str.maketrans('', '', string.punctuation)) for s in data[:,0]]
data[:,1] = [s.translate(str.maketrans('', '', string.punctuation)) for s in data[:,1]]

# convert to lowercase
for i in range(len(data)):
    data[i,0] = data[i,0].lower()
    data[i,1] = data[i,1].lower()

# remove unicode characters   
for i in range(len(data)):
    data[i,0] = data[i,0].replace('\u202f','').strip()
    data[i,1] = data[i,1].replace('\u202f','').strip()
#                                
# get the maximum sentence size and display statistics
def get_max_size(data):
    fr_length = []
    en_length = []
    max_size_sentence_fr = 0
    max_size_sentence_en = 0
    for i in range(len(data)):
        fr_length.append(len(data[i,0].split()))
        if len(data[i,0].split()) > max_size_sentence_fr:
            max_size_sentence_fr = len(data[i,0].split())
        en_length.append(len(data[i,1].split()))
        if len(data[i,1].split()) > max_size_sentence_en:
            max_size_sentence_en = len(data[i,1].split())
    return fr_length, en_length, max_size_sentence_fr, max_size_sentence_en


fr_length, en_length, max_size_sentence_fr, max_size_sentence_en = get_max_size(data)

# display statistics
length_df_fr = pd.DataFrame({'fr':fr_length})
length_df_en = pd.DataFrame({'en':en_length})
length_df = pd.DataFrame({'fr':fr_length, 'en':en_length})

plt.hist(length_df['fr'], bins = 'auto', color = '#b7657b', alpha=0.5, rwidth=0.85, label = 'French')
plt.hist(length_df['en'], bins = 'auto', color = '#78c4d4', alpha=0.5, rwidth=0.85, label = 'English')
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Sentences Lengths')
plt.ylabel('Frequency')
plt.title("Sentence lengths in French and in English")
plt.legend(loc='upper right')
plt.show()
```

```python
# vectorize data by using keras tokenizer class
def tokenizer_function(data):
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(data)
    return tokenizer

tokenizer_fr = tokenizer_function(data[:,0])
size_vocab_fr = len(tokenizer_fr.word_index) + 1

tokenizer_en = tokenizer_function(data[:,1])
size_vocab_en = len(tokenizer_en.word_index) + 1

print("size_vocab_fr: ", size_vocab_fr, " / max_size_sentence_fr: ", max_size_sentence_fr)
print("size_vocab_en: ", size_vocab_en, " / max_size_sentence_en: ", max_size_sentence_en)
```

```python
# split data into training and testing sets (testing set is 10%)
training_data, testing_data = train_test_split(data, test_size=0.1, random_state = 10)
```

```python
# seq2seq model: convert to integer sequences of fixed length
def encoder(tokenizer, max_length, data):
    sequence = tokenizer.texts_to_sequences(data)
    sequence = pad_sequences(sequence, maxlen = max_length, padding = 'post')
    return sequence

X_training = encoder(tokenizer_fr, max_size_sentence_fr, training_data[:, 0])
Y_training = encoder(tokenizer_en, max_size_sentence_en, training_data[:,1])
print(Y_training)
```

# Step 3
### training the model

```python
# record the evolution of the loss at each batch for each epoch
class LossHistory(Callback):
    def on_train_begin(self, logs={}):
        self.loss = []
    def on_epoch_begin(self, batch, logs={}):
        self.tmp = []
    def on_train_batch_end(self, batch, logs={}):
        self.tmp.append(logs["loss"]) 
    def on_epoch_end(self, batch, logs={}):
        self.loss.append(self.tmp)        

# stop training when the loss starts increasing
# waits for "patience" iterations 
class EarlyStopping(Callback):
    def __init__(self, patience=0):
        self.patience = patience
    def on_train_begin(self, logs={}):
        self.best_loss = np.Inf
        self.wait = 0
        self.stopped_epoch = 0
    def on_epoch_end(self, epoch, logs={}):
        current_loss = logs["loss"]
        if current_loss < self.best_loss:
            self.wait = 0
            self.best_loss = current_loss
        else:
            self.wait +=1
            if self.wait > self.patience:
                self.stopped_epoch = epoch
                self.model.stop_training = True
    def on_train_end(self, logs={}):  
        if self.stopped_epoch > 0:
            print("Early stopping at epoch ", str(self.stopped_epoch + 1))

loss_callback = LossHistory()
```

```python
# define the model
# consists of the following layers < embedding , LSTM, LSTM, Dense> 
# size_vocab_fr = number of distinct words in the entire French dataset
# size_vocab_en = number of distinct words in the entire English dataset
# max_size_sentence_fr = size of vectors for French sentences
# max_size_sentence_en = size of vectors for English sentences

# group a linear stack of layers
model = Sequential()
# 512 is the size of the vector space in which the words will be embedded
model.add(Embedding(size_vocab_fr, 512, input_length = max_size_sentence_fr, mask_zero = True))
model.add(LSTM(512))
model.add(RepeatVector(max_size_sentence_en))
model.add(LSTM(512, return_sequences=True))
model.add(Dense(size_vocab_en, activation='softmax'))

rms = optimizers.RMSprop(lr=0.001)
model.compile(optimizer=rms, loss='sparse_categorical_crossentropy', metrics = ['accuracy'])

model.summary()

# train the model
patience = 3
history = model.fit(X_training, Y_training.reshape(np.shape(Y_training)[0], np.shape(Y_training)[1], 1),
                    epochs=90, validation_split = 0.2, 
                    verbose=1, callbacks=[loss_callback, EarlyStopping(patience)])
```

```python
# get the loss for each epoch
loss = loss_callback.loss
```

```python
# display evolution of accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['training set', 'validation set'], loc='upper left')
plt.show()

# display evolution of loss by epoch
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['training set', 'validation set'], loc='upper right')
plt.show()

# display the evolution of loss by batch for every 5 epoch
fontP = FontProperties()
fontP.set_size('x-small')

for epoch_index in range(np.shape(loss)[0]):
    if (epoch_index + 1) == 1 or (epoch_index + 1) % 5 == 0:
        x = np.arange(np.shape(loss)[1]) + 1
        plt.plot(x, loss[epoch_index])
plt.title('Loss for each epoch')
plt.ylabel('Loss')
plt.xlabel('Batch')
legend = [1]
count = 5
while count <= np.shape(loss)[0]:
    legend.append(count)
    count += 5  
plt.legend(['Epoch ' + str(i) for i in legend], bbox_to_anchor=(1.05, 1), loc='upper left', prop=fontP)
plt.show()
```

# Step 4
### Testing the model

```python
def decoder(tokenizer, n):
    for word, value in tokenizer.word_index.items():
        if value == n:
            return word
    return -1

def sequence_to_text(prediction, tokenizer):
    text = []
    for sequence in prediction:
        sentence = ""
        for n in sequence:
            if n != 0:
                word = decoder(tokenizer, n)
                sentence += word
                sentence += " "
        text.append(sentence)
    return text
```

```python
# testing the model on the test set
X_test = encoder(tokenizer_fr, max_size_sentence_fr, testing_data[:, 0])
Y_test = encoder(tokenizer_en, max_size_sentence_en, testing_data[:,1])

prediction = model.predict_classes(X_test.reshape((X_test.shape[0],X_test.shape[1])))
prediction_text = sequence_to_text(prediction, tokenizer_en)
# print(prediction_text)
```

```python
# testing the model on the training set
prediction_training = model.predict_classes(X_training.reshape((X_training.shape[0],X_training.shape[1])))
prediction_text_training = sequence_to_text(prediction_training, tokenizer_en)
```

# Step 5
### Evaluating accuracy

```python
# evaluate accuracy on the test set
pd.set_option('display.max_colwidth', -1)

real_translation = sequence_to_text(Y_test, tokenizer_en)

error_test = wer(real_translation, prediction_text)
print("Error test set: ", error_test)

result = pd.DataFrame({'real_translation': real_translation, 'predicted_translation': prediction_text})
result.sample(10)
```

```python
# evaluate accuracy on the training set
# 1) evaluate function from Keras
loss, accuracy = model.evaluate(X_training, Y_training, verbose=0)
print("Keras accuracy: ", accuracy)

real_translation_training = sequence_to_text(Y_training, tokenizer_en)

error_training = wer(real_translation_training, prediction_text_training)
print("Error training set: ", error_training)

result_training = pd.DataFrame({'real_translation': real_translation_training, 'predicted_translation': prediction_text_training})
result_training.sample(10)
```
